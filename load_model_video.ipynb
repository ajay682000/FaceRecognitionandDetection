## Detecting faces from video or from webcam

```python
import numpy as np
import cv2
import os
```

```python
import import_ipynb
import Face_Recognition as fr
print(fr)
```

```python
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
face_recognizer.read(r'trainingdata.yml') # Give the path to our trained yml file
cap = cv2.VideoCapture(0) 
# 0 indicates the webcam will open
# Replace 0 with the video file path for detecting faces in video
# Replace 0 with 1 for detecting faces from a external camera
name={0:"Person0",1:"Person1",2:"Person2",3:"Person3"}

while True:
    ret,test_img = cap.read()
    faces_detected,gray_img = fr.faceDetection(test_img)
#     print("Faces Detected:", faces_detected)

    for (x,y,w,h) in faces_detected:
        cv2.rectangle(test_img, (x,y), (x+w, y+h), (0,255,0), thickness = 5)
        
    for face in faces_detected:
        (x,y,w,h) = face
        roi_gray = gray_img[x:x+w, y:y+h]
        label, confidence = face_recognizer.predict(roi_gray)
        print("Confidence:", confidence)
        print("Label:", label)
        fr.draw_rect(test_img, face)
        predicted_name = name[label]
        fr.put_text(test_img, predicted_name, x,y)
        
#     resized_img = cv2.resize(test_img, (1000, 700))    
#     cv2.imshow("Face Detection", resized_img)
    
    cv2.imshow("Face Detection", test_img)
    if (cv2.waitKey(1) & 0xFF == ord('q')):
        break
cap.release()
cv2.destroyAllWindows()
```
